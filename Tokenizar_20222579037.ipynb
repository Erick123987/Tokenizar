```python
import pandas as pd
import nltk
from nltk import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import matplotlib.pyplot as plt
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/pycaret/pycaret/master/datasets/amazon.csv') #esta leyendo unos comentarios de 200000
data
```

```python
data["reviewText"].values.tolist() # Los valores son extraídos y transformados en una lista.
```

```python
tokenizar_datos=[] # Se crea una lista para almacenar los datos 
```

```python
#tokenizar los datos de texto de la columna "reviewText" de un conjunto de datos y, luego, imprimir los tokens resultantes.
for text in data["reviewText"]:
  tokens = word_tokenize(text)
  tokenizar_datos.append(tokens)
for tokens in tokenizar_datos: 
  print(tokens)
```

```python
stop_words = stopwords.words('english') #Nos muestra las palabras mas comunes en el idioma seleccionado 
print(stop_words)
```

```python
print(data.reviewText) #imprime las palabras de la lista 
```

```python
#Importación de funciones
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
range_words = range(0, 20000) #Se define el rango de palabras
reviews = data.iloc[range_words]["reviewText"] #selecciona solo las filas del rango y las almacena en review text
```

```python
#Extracción de palabras de token 
words = []
for review in reviews:  
    sentences = sent_tokenize(review) 
    for sentence in sentences:  
        words.extend(word_tokenize(sentence)) 

print(words)
```

```python
#Eliminar las palabras que no son necesarios
words = [w for w in words if not w in stop_words and w.isalpha()] 
print(words)
```

```python
#Creación de wordcloud
from wordcloud import WordCloud, STOPWORDS 
```

```python
# Se realiza la formación de la nube de palabras y se imprime
nube_palabras = ' '.join(words) 
wordcloud = WordCloud( 
    width = 400, 
    height = 800, 
    background_color = 'RED', 
    stopwords = STOPWORDS, 
    relative_scaling = 0, 
    min_font_size = 15 
).generate(nube_palabras) 
 
plt.figure (figsize = (8,8), facecolor = None) 
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis("off") 
plt.tight_layout(pad = 0) 
```

 **CONCLUSIONES**

Los stopwords son palabras comunes que no aportan un valor semántico significativo al análisis de un texto. ya que son muy frecuentes en el lenguaje pero no aportan información relevante sobre el contenido del documento. Por lo tanto, es común eliminar los stopwords durante el procesamiento del texto, ya que su presencia puede afectar negativamente la calidad de los resultados y consumir recursos innecesarios. La eliminación de stopwords permite enfocarse en las palabras más importantes y significativas del texto.


Las nubes de palabras o WordCloud muestran las palabras más frecuentes en un texto de manera gráfica. El tamaño de cada palabra en la nube se determina según su frecuencia en el texto. Las palabras más frecuentes aparecen más grandes en la nube, lo que permite identificar rápidamente los términos clave o los temas principales en un documento. La disposición aleatoria de las palabras en la nube agrega un elemento estético atractivo a la visualización.

La tokenización es un paso fundamental en el procesamiento del lenguaje natural, ya que permite descomponer el texto en unidades significativas para su posterior análisis y procesamiento.
Realizar la tokenización permite dividir el texto en unidades significativas, mientras que la eliminación de stopwords ayuda a filtrar palabras vacías y centrarse en las más relevantes. Estas técnicas son fundamentales para realizar un análisis de texto efectivo y obtener información útil a partir de los documentos analizados.
